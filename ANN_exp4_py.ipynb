```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-10, 10, 400)
def linear(x):
    return x
def relu(x):
    return np.maximum(0, x)
def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
def tanh(x):
    return np.tanh(x)
def bipolar_sigmoid(x):
    return (1 - np.exp(-x)) / (1 + np.exp(-x))
def softplus(x):
    return np.log(1 + np.exp(x))
plt.figure(figsize=(14, 10))
plt.subplot(3, 3, 1)
plt.plot(x, linear(x), label="Linear", color="blue")
plt.title("Linear Activation Function")
plt.grid(True)
plt.subplot(3, 3, 2)
plt.plot(x, relu(x), label="ReLU", color="green")
plt.title("ReLU Activation Function")
plt.grid(True)
plt.subplot(3, 3, 3)
plt.plot(x, leaky_relu(x), label="Leaky ReLU", color="red")
plt.title("Leaky ReLU Activation Function")
plt.grid(True)
plt.subplot(3, 3, 4)
plt.plot(x, sigmoid(x), label="Sigmoid", color="purple")
plt.title("Sigmoid Activation Function")
plt.grid(True)
plt.subplot(3, 3, 5)
plt.plot(x, tanh(x), label="Tanh", color="orange")
plt.title("Hyperbolic Tangent Activation Function")
plt.grid(True)
plt.subplot(3, 3, 6)
plt.plot(x, bipolar_sigmoid(x), label="Bipolar Sigmoid", color="cyan")
plt.title("Bipolar Sigmoid Activation Function")
plt.grid(True)
plt.subplot(3, 3, 7)
plt.plot(x, softplus(x), label="Softplus", color="magenta")
plt.title("Softplus Activation Function")
plt.grid(True)
plt.tight_layout()
plt.show()
```
